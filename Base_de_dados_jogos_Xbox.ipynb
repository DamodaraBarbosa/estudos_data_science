## **Estudos dataset Xbox One Games**

Importando a base de dados

```python
import pandas as pd
import numpy as np
```

```python
data_xbox = pd.read_csv('xbox_one_games.csv')
```

Tratando a base de dados

```python
# a ideia é unir essa base de dados com outra já tratada com jogos que fazem parte do serviço Xbox Gamepass
# as colunas de interesse para esse fim são: 'name', 'publisher', 'developer', 'release', 'platform', 'genre', 'notes', 
# 'medium', 'size', 'features'
```

```python
# quanto a coluna 'publisher', percebe-se que em alguns casos há mais de uma publicadora, a ideia é fragmentar tais casos
# para uma futura análise

data_xbox[['publisher']].sample(10)
```

```python
# definindo a nova coluna 'publisher'

data_xbox['publisher'] = data_xbox['publisher'].str.split(', ')
data_xbox.head()
```

```python
# a ideia para coluna 'developer' é a mesma, pois um mesmo jogo pode ser desenvolvido em conjunto por mais de um estúdio

data_xbox['developer'] = data_xbox['developer'].str.split(', ')
data_xbox.head()
```

```python
# para os dados da coluna 'release' com valor 'To be announced'

data_xbox.query('release == "To be announced"')
# data_xbox['release'] = pd.to_datetime(data_xbox['release'])
# data_xbox.head()
```

```python
# após uma breve pesquisa é possível substituir os valores
# Operation Zeta: https://www.xbox.com/pt-BR/games/store/operation-zeta/9P00HTX0XXRV/
# Demoniaca: Everlasting Night: https://www.xbox.com/pt-br/games/store/demoniaca-everlasting-night/9pm3q3dx71p2/
# Kraken Academy!!: https://www.xbox.com/pt-br/games/store/kraken-academy/9p87cgf1tccx/

data_xbox.at[1327, 'release'] = '13 Jan 2022'
data_xbox.at[2635, 'release'] = '11 Jan 2022'
data_xbox.at[3492, 'release'] = '21 Mar 2022'

data_xbox.loc[[1327, 2635, 3492]]
```

```python
# converte-se a coluna 'release' para datetime

data_xbox['release'] = pd.to_datetime(data_xbox['release'])
data_xbox.head()
```

```python
data_xbox.info()
```

```python
# fragmenta-se os valores da coluna 'platform'

data_xbox['platform'] = data_xbox['platform'].str.split(', ')
data_xbox.head()
```

```python
# fragmenta-se os valores da coluna 'genre'

data_xbox['genre'] = data_xbox['genre'].str.split(', ')
```

```python
# fragmenta-se os valores da coluna 'notes'

data_xbox['notes'] = data_xbox['notes'].str.split(', ')
```

```python
# para coluna 'size' é necessário distanciar o valor numérico da unidade de armazenamento de dados
# mas para facilitar a iteração substitui-se os valores nulos por 0

data_xbox['size'] = data_xbox['size'].fillna('0MB')
data_xbox.query('size == "0MB"').sample(5)
```

```python
size_list = list()

for index, size in enumerate(data_xbox['size']):
  for posicao, caractere in enumerate(size):
    if size[posicao].isalpha() == True and size[posicao-1].isdigit() == True:
      size = list(size)
      size.insert(posicao, ' ')
      size_list.append(''.join(size))

size_list[:10]
```

```python
# os valores foram organizados na colunas 'size'

data_xbox['size'] = size_list
len(data_xbox['size'])
```

```python
size_games = data_xbox['size'].str.split(expand = True)
size_games.head()
```

```python
# apaga-se a 'velha' coluna 'size' e define-se duas novas colunas com o valor numérico do tamanho do arquivo e a unidade de armazenamento

data_xbox.pop('size')
data_xbox['size'] = size_games[0]
data_xbox['unit'] = size_games[1]
data_xbox.head(2)
```

```python
# converte-se a coluna 'size' de object para float

data_xbox['size'] = data_xbox['size'].astype(float)
data_xbox.info()
```

```python
# também converte-se os valores para unidade de armazenamento GB, de forma a padronizar as colunas

armazenamento_gb = list()


for index, unidade in enumerate(data_xbox['unit']):
  if unidade == 'MB':
    armazenamento_em_gb = float(data_xbox['size'][index])/1000
    armazenamento_gb.append(armazenamento_em_gb)

len(armazenamento_gb)
```

```python
contador = 0 

for index, unidade in enumerate(data_xbox['unit']):
  if unidade == 'MB':
    data_xbox['size'][index] = armazenamento_gb[contador]
    contador += 1

data_xbox.head()
```

```python
data_xbox.columns
```

```python
#revome-se a coluna 'unit' e renomeia-se a coluna 'size'

data_xbox.pop('unit')
data_xbox.columns = ['gameid', 'name', 'web', 'publisher', 'developer', 'release', 'platform', 'genre', 'hardware', 'notes', 
                     'medium', 'completion est', 'links', 'features', 'size (GB)']
data_xbox.head()
```

```python
# quanto a coluna 'medium', os valores existentes 

data_xbox['medium'].unique()
```

```python
# estabelecendo os valores 'Digital', 'Physical' e 'Digital, Physical'

contador = 0

for index, medium in enumerate(data_xbox['medium']):
  try:
    if medium == 'Digital only':
      data_xbox['medium'][index] = 'Digital'
    elif medium == 'Physical only':
      data_xbox['medium'][index] = 'Physical'
    elif medium == None:
      pass
    else:
      data_xbox['medium'][index] = 'Digital, Physical'
  except TypeError:
    pass
```

```python
# separando os valores quando este for composto por mais de um elemento

data_xbox['medium'] = data_xbox['medium'].str.split(', ')
data_xbox.sample(20)
```

```python
# quanto a coluna 'completion est' como há um intervalo, uma estimativa do tempo para completar o jogo, é necessário 
# um tratamento para calcular a média de tempo 

data_xbox['average time (h)'] = data_xbox['completion est'].str.replace('-', ' ')
data_xbox.pop('completion est')
```

```python
data_xbox.head()
```

```python
data_xbox['average time (h)'].fillna('0 hours', inplace = True)
```

```python
data_xbox['average time (h)'].isna().unique()
```

```python
data_xbox.head()
```

```python
# separa-se os dados da nova coluna 'average time (h)' para determinar a média entre o intervalo de horas

average_time = data_xbox['average time (h)'].str.split(expand = True)
average_time.head()
```

```python
average_time[0].unique()
```

```python
average_time[1].unique()
```

```python
# estabelecendo tempo médio para concluir o jogo

for index, times in enumerate(data_xbox['average time (h)']):
  if '+' in average_time[0][index]:
    ate_penultimo_caractere = len(average_time[0][index])-1
    data_xbox.at[index, 'average time (h)'] = float(average_time[0][index][:ate_penultimo_caractere])
  elif 'hours' in average_time[1][index]:
    data_xbox.at[index, 'average time (h)'] = float(average_time[0][index])
  else:
    data_xbox.at[index, 'average time (h)'] = (float(average_time[1][index]) + float(average_time[0][index]))/2

data_xbox.head()
```

```python
data_xbox['average time (h)'].unique()
```

```python
# para finalizar, a coluna 'features' deve ser separada

data_xbox['features'] = data_xbox['features'].str.split(', ')
data_xbox.head()
```

Exportando a base de dados tratada
