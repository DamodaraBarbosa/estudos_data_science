# Tratamento da tabela com os jogos do servi√ßo Xbox Gamepass

```python
import pandas as pd
```

Importando a base de dados

```python
gamepass_data = pd.read_csv('Game Pass Master List - Master List.csv')
```

Tratando a base de dados

```python
# renomeando as colunas da base de dados

columns = list()

for key, value in gamepass_data.iloc[0].items():
  columns.append(value)

columns
```

```python
gamepass_data.columns = columns
```

```python
gamepass_data.drop(0, inplace = True)
```

```python
gamepass_data.head()
```

```python
# convertendo os dados da coluna 'Added' para datetime
gamepass_data['Added'] = pd.to_datetime(gamepass_data['Added'])
gamepass_data.info()
```

```python
# convertendo a coluna 'Release' de object para datetime

gamepass_data['Release'] = pd.to_datetime(gamepass_data['Release'])
```

```python
gamepass_data.sample(5)
```

```python
# convertendo a coluna 'Removed' para datetime

gamepass_data['Removed'] = pd.to_datetime(gamepass_data['Removed'])
```

```python
gamepass_data.sample(5)
```

```python
# convertendo a coluna 'Metacritic' de object para int

for index, metacritic in enumerate(gamepass_data['Metacritic']):
  try:
    gamepass_data.at[index, 'Metacritic'] = float(metacritic)
  except:
    gamepass_data.at[index, 'Metacritic'] = float(metacritic)
    pass
```

```python
# os g√™neros de jogos que aparecem na coluna 'Genre (Giantbomb)'

gamepass_data['Genre (Giantbomb)'].unique()
```

```python
# substituindo o elemento 'Role-Playing' para RPG na coluna 'Genre (Giantbomb)'

gamepass_data['Genre (Giantbomb)'] = gamepass_data['Genre (Giantbomb)'].str.replace('Role-Playing', 'RPG')
```

```python
# substituindo o elemento 'First-Person Shooter' para FPS na coluna 'Genre (Giantbomb)'

gamepass_data['Genre (Giantbomb)'] = gamepass_data['Genre (Giantbomb)'].str.replace('First-Person Shooter', 'FPS')
```

```python
# separando os elementos dos dados da coluna 'Genre (Giantbomb)'

gamepass_data['Genre (Giantbomb)'] = gamepass_data['Genre (Giantbomb)'].str.split(' / ')
```
