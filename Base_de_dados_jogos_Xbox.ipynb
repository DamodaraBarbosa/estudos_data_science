## **Estudos dataset Xbox One Games**

Importando a base de dados

```python
import pandas as pd
import numpy as np
```

```python
data_xbox = pd.read_csv('xbox_one_games.csv')
```

```python
data_xbox.head()
```

```python
data_xbox.info()
```

Tratando a base de dados

```python
# a ideia é unir essa base de dados com outra já tratada com jogos que fazem parte do serviço Xbox Gamepass
# as colunas de interesse para esse fim são: 'name', 'publisher', 'developer', 'release', 'platform', 'genre', 'notes', 
# 'medium', 'size', 'features'
```

```python
# quanto a coluna 'publisher', percebe-se que em alguns casos há mais de uma publicadora, a ideia é fragmentar tais casos
# para uma futura análise

data_xbox[['publisher']].sample(10)
```

```python
# definindo a nova coluna 'publisher'

data_xbox['publisher'] = data_xbox['publisher'].str.split(', ')
data_xbox.head()
```

```python
# a ideia para coluna 'developer' é a mesma, pois um mesmo jogo pode ser desenvolvido em conjunto por mais de um estúdio

data_xbox['developer'] = data_xbox['developer'].str.split(', ')
data_xbox.head()
```

```python
# quanto a coluna 'release' referente a data de lançamento dos jogos, que é do tipo object, pode ser convertida para datetime
# alguns dados da coluna 'release' estão com valor 'Yesterday', são os mostrados abaixo

data_xbox.query('release == "Yesterday"')
```

```python
# com uma rápida pesquisa é possível definir a data de lançamento para os jogos acima:
# Winter Sports Games: https://store.steampowered.com/app/998530/Winter_Sports_Games/
# Gardener's Path: https://www.ratalaikagames.com/games.php?id=gardenerspath/
# Lacuna - Sci-Fi Noir Adventure: https://www.xbox.com/en-US/games/store/lacuna-a-sci-fi-noir-adventure/9MXKTG2JC5SN/

data_xbox.at[136, 'release'] = '24 Dec 2019'
data_xbox.at[3053, 'release'] = '22 Dec 2021'
data_xbox.at[3505, 'release'] = '20 Dec 2021'
```

```python
# verificando se os dados foram modificados corretamente

data_xbox.loc[[136, 3053, 3505]]
```

```python
# ainda há a presença de valores como 'Today' e 'Tomorrow' que não são de interesse

for release in data_xbox['release'].str.split():
  try:
    if len(release) == 1:
      print(release)
  except:
    pass
```

```python
# localizando-os no dataframe

data_xbox.query('release == "Today" or release == "Tomorrow"')
```

```python
# substiui-se tais valores pelos corretos obtidos por pesquisas
# In My Shadow: https://www.xbox.com/pt-br/games/store/In-My-Shadow/9NSRQQGSFP1N/
# Planet Quiz: Learn & Discover: https://www.xbox.com/pt-BR/games/store/planet-quiz-learn-discover/9N4GZJJR1M84/
# Horatio Goes Snowboarding: https://www.xbox.com/pt-BR/games/store/horatio-goes-snowboarding/9PGBGJQ4FDGC/0010/
# Just a Phrase by POWGI:https://www.xbox.com/pt-br/games/store/just-a-phrase-by-powgi/9pnh58mkp48n/

data_xbox.at[1194, 'release'] = '21 Dec 2021'
data_xbox.at[1492, 'release'] = '22 Dec 2021'
data_xbox.at[3314, 'release'] = '21 Dec 2021'
data_xbox.at[3392, 'release'] = '21 Dec 2021'

data_xbox.loc[[1194, 1492, 3314, 3392]]
```

```python
# para os dados da coluna 'release' com valor 'To be announced'

data_xbox.query('release == "To be announced"')
# data_xbox['release'] = pd.to_datetime(data_xbox['release'])
# data_xbox.head()
```

```python
# após uma breve pesquisa é possível substituir os valores
# Operation Zeta: https://www.xbox.com/pt-BR/games/store/operation-zeta/9P00HTX0XXRV/
# Demoniaca: Everlasting Night: https://www.xbox.com/pt-br/games/store/demoniaca-everlasting-night/9pm3q3dx71p2/
# Kraken Academy!!: https://www.xbox.com/pt-br/games/store/kraken-academy/9p87cgf1tccx/

data_xbox.at[1327, 'release'] = '13 Jan 2022'
data_xbox.at[2635, 'release'] = '11 Jan 2022'
data_xbox.at[3492, 'release'] = '21 Mar 2022'

data_xbox.loc[[1327, 2635, 3492]]
```

```python
# converte-se a coluna 'release' para datetime

data_xbox['release'] = pd.to_datetime(data_xbox['release'])
data_xbox.head()
```

```python
data_xbox.info()
```

```python
# fragmenta-se os valores da coluna 'platform'

data_xbox['platform'] = data_xbox['platform'].str.split(', ')
data_xbox.head()
```

```python
# fragmenta-se os valores da coluna 'genre'

data_xbox['genre'] = data_xbox['genre'].str.split(', ')
```

```python
# fragmenta-se os valores da coluna 'notes'

data_xbox['notes'] = data_xbox['notes'].str.split(', ')
```

```python
# para coluna 'size' é necessário distanciar o valor numérico da unidade de armazenamento de dados
# mas para facilitar a iteração substitui-se os valores nulos por 0

data_xbox['size'] = data_xbox['size'].fillna('0MB')
data_xbox.query('size == "0MB"').sample(5)
```

```python
size_list = list()

for index, size in enumerate(data_xbox['size']):
  for posicao, caractere in enumerate(size):
    if size[posicao].isalpha() == True and size[posicao-1].isdigit() == True:
      size = list(size)
      size.insert(posicao, ' ')
      size_list.append(''.join(size))

size_list
```

```python
# os valores foram organizados na colunas 'size'

data_xbox['size'] = size_list
data_xbox['size'].sample(20)
```

```python
size_games = data_xbox['size'].str.split(expand = True)
size_games.head()
```

```python
# apaga-se a 'velha' coluna 'size' e define-se duas novas colunas com o valor numérico do tamanho do arquivo e a unidade de armazenamento

data_xbox.pop('size')
data_xbox['size'] = size_games[0]
data_xbox['unit'] = size_games[1]
data_xbox.head(2)
```

```python
# converte-se a coluna 'size' de object para float

data_xbox['size'] = data_xbox['size'].astype(float)
data_xbox.info()
```
