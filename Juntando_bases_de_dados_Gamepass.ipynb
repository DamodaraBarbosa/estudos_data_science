# Unindo os dados de duas bases diferentes
## Quanto aos jogos do Xbox e Xbox Gamepass

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 1000)
```

Importanto as bases de dados

```python
gamepass = pd.read_csv('/content/data_gamepass_tratado.csv')
xbox = pd.read_csv('/content/xbox_one_games_tratada.csv')
tabela_gamepass = pd.read_csv('gamepass_data_tratado.csv')
```

Unificando os dataframes gamepass e tabela_gamepass

```python
# primeiro o merge é feito com os dataframes tabela_gamepass e gamepass
```

```python
dados_gamepass = pd.merge(left = tabela_gamepass, right = gamepass[['GAME', 'RATIO', 'GAMERS', 'COMP %', 'AVERAGE TIME (h)', 'RATING', 'True_Achievement', 'Game_Score']], how = 'left', left_on= 'Game', right_on= 'GAME')
```

```python
dados_gamepass.sample(10)
```

```python
dados_gamepass.shape
```

```python
# junta-se também o dataframe xbox
```

```python
dados_gamepass = pd.merge(left= dados_gamepass, right= xbox[['name', 'publisher', 'developer', 'platform', 'hardware', 'notes', 'medium', 'features', 'size (GB)']], how= 'left', left_on= 'Game', right_on= 'name')
```

```python
dados_gamepass.shape
```

```python
dados_gamepass.sample(10)
```

Exportanto a base de dados 

```python
gamepass_database = dados_gamepass.to_csv('mega_basededados_gp.csv', index = False)
```
