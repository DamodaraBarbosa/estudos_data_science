Tratamento da base de dados

```python
data_gamepass.info()
```

```python
# a coluna 'RATIO' aparece como dtype: object, apesar de haver floats nos seus dados
# é necessário, então, converter os valores, porém visualizando os valores únicos percebe-se que 

data_gamepass['RATIO'].unique()
```

```python
# há o valor '-' para ratio, é possível localizá-lo no dataframe

data_gamepass.query('RATIO == "-"')
```

```python
# três valores que aparecem com '-' como valor de 'RATIO'
# modifica-se tal valor para None

data_gamepass['RATIO'].loc[[452, 453, 454]] = None
```

```python
# converte-se os valores de 'RATIO' para float 
data_gamepass['RATIO'] = data_gamepass['RATIO'].str.replace('.', '.').astype(float)
```

```python
# preenche-se os dados de 'RATIO' nulos com a mediana da coluna de maneira definitiva

data_gamepass['RATIO'].fillna(data_gamepass['RATIO'].median(), inplace = True)
```

```python
# para tratar a coluna 'ADDED' referente a data de entrada do jogo no Gamepass

data_gamepass['ADDED'].unique()
```

```python
# percebe-se a presença do valor 'Yesterday' que pode ser localizado usando a função .query()

data_gamepass.query('ADDED == "Yesterday"')
```

```python
# pesquisando a data de entrada do jogo no serviço de jogos da Microsoft verifica-se que este entrou no dia 02 de maio de 2022 
# fonte: https://news.xbox.com/en-us/2022/05/02/loot-river-is-now-available-for-pc-xbox-one-and-xbox-series-xs-game-pass/
# então substitui-se o valor para o correto

data_gamepass.at[438, 'ADDED'] = '02 May 22'
```

```python
data_gamepass.loc[438]
```

```python
# os valores da coluna 'ADDED' são do tipo object, porém deseja-se que estes estejam em datetime, sendo necessário a sua conversão

data_gamepass['ADDED'] = pd.to_datetime(data_gamepass['ADDED'])
```

```python
data_gamepass.info()
```

```python
data_gamepass.head()
```

```python
data_gamepass['TIME'].unique()
```

```python
# há um valor na coluna 'TIME' 1000+ hours, é possível identificá-lo na base de dados

data_gamepass.query('TIME == "1000+ hours"')
```

```python
data_gamepass['TIME'].loc[[26, 37, 109, 180, 214]] = '1000-'
```

```python
# a coluna 'TIME' traz uma estimativa do tempo para completar o jogo, a ideia é transformar cada valor na média dos intervalos, para isso:

data_gamepass_time = data_gamepass['TIME'].str.split(expand = True)
```

```python
data_gamepass_time.head()
```

```python
data_gamepass['TIME'] = data_gamepass_time[0]
data_gamepass.head()
```

```python
# substitui-se o caractere '-' por ', ' para os dados na coluna 'TIME'

data_gamepass['TIME'] = data_gamepass['TIME'].str.replace('-', ' ')
data_gamepass.head()
```

```python
# utiliza-se o split para separar os dois intervalos de tempo

data_gamepass['TIME'] = data_gamepass['TIME'].str.split()
```

```python
data_gamepass.head()
```

```python
data_gamepass['TIME'].loc[[26, 37, 109, 180, 214]]
```

```python
# define-se uma variável do tipo np.array para definir a média das horas para completar os jogos, quando o len() do dado for apenas 1 não é necessário fazer a média

times = np.array(data_gamepass['TIME'])

for index, time in enumerate(times):
  try:
    if len(time) == 2:
      times[index] = ((float(time[0]) + float(time[1]))/len(time))
    elif type(time) == float:
      times[index] = time
    elif time == list(['1000']):
      times[index] = 1000.0
  except TypeError:
    pass

times
```

```python
# comparando o tamanho dos dados para garantir que nenhum se perdeu

print(len(times))
print(len(data_gamepass['TIME']))
```

```python
data_gamepass['AVERAGE TIME (h)'] = times
data_gamepass.head()
```

```python
# retirando a coluna 'TIME' e organizando as colunas do dataframe
data_gamepass.pop('TIME')
data_gamepass = data_gamepass[['GAME', 'RATIO', 'GAMERS', 'COMP %', 'AVERAGE TIME (h)', 'RATING', 'ADDED', 'True_Achievement', 'Game_Score']]
data_gamepass.head()
```

```python
data_gamepass['GAMERS'] = data_gamepass['GAMERS'].str.replace(',', '.').astype(float)
data_gamepass.head()
```
