Importando a biblioteca Pandas

```python
import pandas as pd
```

# Definindo os dataframes com jogos Xbox

```python
# dataframe com biblioteca de jogos do Xbox One:

df_xbox_one = pd.read_csv('/content/drive/MyDrive/Dados Xbox/df_xbox_games.csv')
df_xbox_one.head()
```

```python
df_xbox_one.shape
```

```python
# dataframe com bilbioteca de jogos do Xbox Series:

df_xbox_series = pd.read_csv('/content/drive/MyDrive/Dados Xbox/xbox_series_games.csv')
df_xbox_series.head()
```

```python
df_xbox_series.shape
```

# Verificando se há jogos que se repetem em ambas as bibliotecas

```python
# definem-se as variáveis que armazenam os jogos das plataformas Xbox:

xbox_one_games = list(df_xbox_one['Name'])
xbox_series_games = list(df_xbox_series['name'])
```

```python
# jogos que se repetem em ambas as

contador = 0

for series in xbox_series_games:
  if series in xbox_one_games:
    print(series)
    contador += 1
```

```python
# quantidade de jogos que se repetem nas bibliotecas:

print(contador)
```

# Concatenação dos dataframes

```python
# um pequeno tratamento necessário é igualar as colunas de ambos os dataframes:

df_xbox_one.columns = ['name', 'genre(s)', 'developer(s)', 'publisher(se)', 'releasedate', 'addons']
```

```python
# concatenando os dataframes:

df_xbox_games = pd.concat([df_xbox_one, df_xbox_series])
```

```python
df_xbox_games.shape
```

Para apagar os valores que se repetem

```python
df_xbox_games.drop_duplicates(subset= 'name', inplace= True)
```

```python
# reestabelecendo o index do dataframe:

df_xbox_games.index = range(df_xbox_games.shape[0])

df_xbox_games
```

Ordenando o dataframe pelo nome do jogo

```python
df_xbox_games.sort_values(by= 'name', inplace= True)
```

```python
df_xbox_games
```

# Exportando o dataframe

```python
df_xbox_games.to_csv('df_xbox_games.csv', index= False)
```
